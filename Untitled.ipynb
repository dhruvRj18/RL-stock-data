```python
class Environment1:
    
    def __init__(self, data, history_t=90):
        self.data = data
        self.history_t = history_t
        self.reset()
        
    def reset(self):
        self.t = 0
        self.done = False
        self.profits = 0
        self.positions = []
        self.position_value = 0
        self.history = [0 for _ in range(self.history_t)]
        return [self.position_value] + self.history # obs
    
    def step(self, act):
        if self.t < len(self.data):
            self.position_value = 0
            for p in self.positions:
                self.position_value += (self.data.iloc[self.t, :]['Close'] - p)
            self.history.pop(0)
            self.history.append(self.data.iloc[self.t, :]['Close'] - self.data.iloc[(self.t-1), :]['Close'])
            # print("self.t:", self.t)
            # print("Length of self.data:", len(self.data))
            self.t += 1
            done = False
            if self.t == len(self.data):
                done = True
            return self.history, self.position_value, done
        else:
            # Handle the case where self.t exceeds the length of self.data
            return None, None, True
```

```python
env = Environment1(train)
observation = env.reset()
print("Initial observation:", observation)

for _ in range(3):
    pact = np.random.randint(3)
    observation, reward, done = env.step(pact)
    print("Action:", pact)
    print("Observation:", observation)
    print("Reward:", reward)
    print("Episode done:", done)
```

```python
class QLearningAgent:
    def __init__(self, actions, epsilon=0.1, alpha=0.2, gamma=0.9):
        self.q_table = dict()
        self.actions = actions
        self.epsilon = epsilon
        self.alpha = alpha
        self.gamma = gamma

    def choose_action(self, state):
        if np.random.uniform(0, 1) < self.epsilon:
            action = np.random.choice(self.actions)
        else:
            if tuple(state) not in self.q_table:
                self.q_table[tuple(state)] = np.zeros(len(self.actions))
            action = self.actions[np.argmax(self.q_table[tuple(state)])]
        return action

    def update_q_table(self, state, action, reward, next_state):
        if tuple(state) not in self.q_table:
            self.q_table[tuple(state)] = np.zeros(len(self.actions))
        if tuple(next_state) not in self.q_table:
            self.q_table[tuple(next_state)] = np.zeros(len(self.actions))
        best_next_action = np.argmax(self.q_table[tuple(next_state)])
        td_target = reward + self.gamma * self.q_table[tuple(next_state)][best_next_action]
        td_error = td_target - self.q_table[tuple(state)][action]
        self.q_table[tuple(state)][action] += self.alpha * td_error

def train_agent(env, num_episodes=1000):
    actions = [0, 1, 2]  # 0: stay, 1: buy, 2: sell
    agent = QLearningAgent(actions)
    
    rewards = []
    losses = []
    
    for episode in range(num_episodes):
        state = env.reset()
        done = False
        episode_reward = 0
        
        while not done:
            action = agent.choose_action(state)
            next_state, reward, done = env.step(action)
            loss = agent.update_q_table(state, action, reward, next_state)
            episode_reward += reward
            state = next_state
            
            # Print Q-values
            print("Q-values:", agent.q_table)
            # Print loss
            print("Loss:", loss)
            # Print reward for each step
            print("Reward:", reward)
            
        rewards.append(episode_reward)
        losses.append(loss)
        
        print("Episode:", episode + 1, "| Total Reward:", episode_reward)
        
    return agent, rewards, losses


def evaluate_agent(agent, env):
    state = env.reset()
    done = False
    total_reward = 0
    while not done:
        action = agent.choose_action(state)
        next_state, reward, done = env.step(action)
        total_reward += reward
        state = next_state
    return total_reward



env = Environment1(train)
trained_agent = train_agent(env)
average_reward = evaluate_agent(trained_agent, env)
print("Average reward:", average_reward)
```
